```python
#3 Create a data frame having at least 3 columns and 50 rows to store numeric data generated using a random function. 
#Replace 10% of the values by null values whose index positions are generated using random function.
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(0,100,size=(50, 3)), columns=list('ABC'))
```

```python
df
```

```python
for col in df.columns:
    df.loc[df.sample(frac=0.1).index, col] = np.nan
```

```python
df
```

```python
#a. Identify and count missing values in a data frame.
missing_values = df.isnull() 
missing_values
```

```python
missing_count = missing_values.sum()
missing_count
```

```python
df.isna()
```

```python
#b. Drop the column having more than 5 null values.
df.dropna(thresh=len(df) - 5, axis=1) 
```

```python
#c. Identify the row label having maximum of the sum of all values in a row and drop that row.
# calculate the row sums and find the index of the row with the maximum sum
max_sum_row_idx = df.sum(axis=1).idxmax()

# find the row label with the maximum sum of values in a row
max_sum_row_label = df.sum(axis=1).idxmax()

# print the row label and its sum
print("Row label with maximum sum of values:", max_sum_row_label)
print("Sum of values in the row:", df.sum(axis=1)[max_sum_row_label])

# drop the row with the maximum sum of values
df = df.drop(index=max_sum_row_label)

# print the resulting data frame
print(df)
```

```python
#e. remove all the duplicates from the first column
df.drop_duplicates(subset='A', inplace=True)
df
```

```python
#f.Find the correlation between first and second column and covariance between second and third column.
corr = df['A'].corr(df['B'])
print(f"\nCorrelation between A and B: {corr}")

# calculate the covariance between the second and third column
cov = df['B'].cov(df['B'])
print(f"Covariance between B and C: {cov}")
```

```python
#g. Discretize the second column and create 5 bins.
pd.cut(df['B'],
       bins=[0, 2.5, 3, 3.25, 3.5, 3.75, 4], 
       labels=["Very bad", "Bad", "Average", "good", "Very good", "Excellent"])
```
